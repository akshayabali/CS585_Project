```python
modified_a = [[0, 0, 0]]
count = 0
for _a in a:
    _x = _a[0]
    _y = _a[1]
    _z = _a[2]

    x = modified_a[count][0]
    y = modified_a[count][1]
    modified_a.append([x + _x, y + _y, _z])
    count += 1
```

```python
def plot(arr):
    a_x = arr[:,0]
    a_y = arr[:,1]
    plt.plot(a_x, a_y, 'ro')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('My chart')
    plt.show()
```

```python

modified_a = np.array(modified_a)
plot(a)
plot(modified_a)
```
